```python
import pandas as pd
df = pd.read_csv('/content/horse_racing.csv')
```

```python
df.info()
```

```python
# 1 Calculate the average weight of jockeys across all rows in the dataset. The weights are given in kilograms. Round the result to two decimal places.

round(df['Jockey weight'].mean(), 2)
```

```python
# 2 What was the average race time for jockey Brett Prebble? Round to two decimal places.

print('Avg Race Time', round(df[df['Jockey'] == 'Brett Prebble']['Race time'].mean(), 2))
```

```python
# 3 Create a new DataFrame containing only the rows where horses and their jockeys won first place. Calculate the average prize money won by first-place jockeys, in Hong Kong Dollars (HKD). Round the result to two decimal places.

Place = df['Final place'] == 1
df_Place = df[Place]
round(df_Place['Prize money'].mean(),2)
```

```python
df.head()
```

```python
# 4 Did winning horses tend to run on dirt or grass surfaces?

df.groupby('Surface')['Surface'].count()
```

```python
#  5 What was the fastest race time among horses that ran the 1400-meter race? The units given are in seconds (sometimes, the data is dirty, just assume they are all seconds for now).

distance = df['Distance'] == 1400
df_distance = df[distance]
df_distance
df_distance.sort_values(by='Race time')
```

```python
# 6 What was the slowest race time among Irish horses who ran the 1400-meter race?

distance_irish = (df['Distance'] == 1400) & (df['Country'] == 'Irland')
df_distance_irish = df[distance_irish]
df_distance_irish
df_distance_irish.sort_values(by='Race time', ascending=False)

# 9291
```

```python
# 7 Which Australian-born horse's jockey has competed in the most races?

australian_horses = df[df['Country'] == 'Australia']
australian_horses
Jockey_counts = australian_horses.groupby('JockeyId')['Jockey'].value_counts()
Jockey_counts.sort_values(ascending=False)
```

```python
# 8 Assuming that higher odds indicate a lower likelihood of winning a race, determine which jockey, riding horses originating from Australia, Ireland, or New Zealand, has the lowest average likelihood of winning (i.e., the highest average odds) based on this dataset.

AusIreNZea = df[(df['Country'] == 'Australia') | (df['Country'] == 'Ireland') | (df['Country'] == 'Australia')]
AvgOdds = AusIreNZea.groupby('Jockey')['Odds'].mean()
AvgOdds.sort_values(ascending=False)
```

```python
# 9 For horses aged 10 or above, identify the horse ID with the best average race time (i.e., the smallest average race time) in the dataset. Answer to the nearest second.

HorseAge = df[df['Horse age'] >= 10]
BestAvg = HorseAge.groupby('HorseId')['Race time'].mean()
BestAvg.sort_values()
```

```python
# 10 Which country, based on the horses' country of origin, achieved the most first-place wins in races where the prize money exceeded $2,000,000?

df[(df['Final place']==1) & (df['Prize money']>2000000)]['Country'].value_counts()
```
